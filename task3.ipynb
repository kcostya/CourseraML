```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline
```

# Задание: центральная предельная теорема своими руками

## Выбранное распределение

В качестве распределения исходных величин взято распределение Парето: https://en.wikipedia.org/wiki/Pareto_distribution

```python
#Строю график плотности распределения
pareto = st.pareto(3,  scale=1)
x = np.linspace(-7,7,1000)
pdf = pareto.pdf(x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title(u'Плотность распределения (Парето)')
print ""
```

```python
#Строю график функции распределения
cdf = pareto.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title(u'Функция распределения (Парето)')
print ""
```

## Генерация выборки величин с распределением Парето

```python
sample = pareto.rvs(1000)
plt.hist(sample, normed=True, bins=100, label=u'Гистограмма выборки')
plt.plot(x, pdf, label=u'Теоретическая плотность')
plt.legend()
plt.xlim(0, 6)
print ""
```

## Оценка распределения выборочного среднего случайной величины при разных объёмах выборок

```python
n = 5
#Генерирую массив из средних значений 1000 выборок по n элементов
means = np.array([np.array(pareto.rvs(n)).mean() for i in range(1000)])
plt.hist(means, normed=True, bins=50, label=u'Гистограмма среднего по выборке')
#Получаю параметры нужного распределения через .stats()
#Дополнительные выкладки по вычислению характеристик распределения не требуются
normal = st.norm(pareto.stats()[0], np.sqrt(pareto.stats()[1]/float(n)))
x = np.linspace(-1,10,1000)
pdf = normal.pdf(x)
plt.plot(x, pdf, label=u'Теоретическая плотность (в соотв с ЦПТ)')
plt.xlim(0, 5)
plt.legend()
print ""
```
